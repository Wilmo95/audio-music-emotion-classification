```python
import pandas as pd
import numpy as np
import requests
```

```python
csv = "http://www2.projects.science.uu.nl/memotion/emotifydata/data.csv"
response = requests.get(csv)
open("data.csv", "wb").write(response.content)
```

```python
data_csv = pd.read_csv('data.csv', skipinitialspace=True)
data_csv
```

```python
genre = data_csv[['track id', 'genre']]
genre
```

```python
labels = data_csv.drop(columns= ['mood', 'age', 'gender','disliked', 'liked', 'mother tongue', 'genre']) # unnecessary columns
labels = labels.groupby('track id').mean()
labels
```

```python
for column in labels.columns:
    labels[column] = labels[column].apply(lambda x: 1 if (x>=0.5) else 0)
    print(labels[column].value_counts())
    print()
```

```python
labels
```

```python
genre_cleared = genre.drop_duplicates().set_index('track id')
genre_cleared
```

```python
y = labels
y['genre'] = genre_cleared['genre']
y = pd.get_dummies(y)
y
```
